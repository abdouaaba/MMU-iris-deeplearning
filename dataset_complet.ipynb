```python
"""===================================================
    Chargement de la base de données "iris"
==================================================="""
from sklearn.datasets import load_iris
donnees_iris = load_iris()
print(donnees_iris.filename) # Chemin vers le fichier de la base de données iris
print(donnees_iris.keys()) # Contenu du fichier de la base iris
print(donnees_iris.DESCR) # Description de la base de données iris
print(donnees_iris['DESCR']) # Équivalente à la ligne précédente
print(donnees_iris['data'][:3]) # Afficher les 3 première lignes de la table des données iris
```

```python
"""===================================================
    Création d'une base de données en format csv
==================================================="""
nbrObj = 150
nbrParam = 4
nomsCls = ['cls1', 'cls2', 'cls3']
fichierOut = open('test.csv', 'w')
fichierOut.write(str(nbrObj))
fichierOut.write(',')
fichierOut.write(str(nbrParam))
for i in range(3):
    fichierOut.write(',')
    fichierOut.write(nomsCls[i])
fichierOut.write('\n')
for i in range(nbrObj):
    for j in range(nbrParam):
        fichierOut.write(str(donnees_iris['data'][i][j]))
        fichierOut.write(',')
        fichierOut.write(str(donnees_iris['target'][i]))
    fichierOut.write('\n')
fichierOut.close()
```
