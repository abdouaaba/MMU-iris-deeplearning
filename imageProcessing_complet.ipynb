```python
"""===============================================
    PIL
==============================================="""
# Déjà installé
from PIL import Image
image4 = Image.open('test.jpeg')
print(type(image4))
image4.show() # l'affichage se fait avec visionneur d'image du système d'exploitation
print('Format de l\'image: ', image4.format)
print('Mode de couleurs: ', image4.mode)
print('Largeur de l\'image: ', image4.width)
print('Hauteur de l\'image: ', image4.height)
plt.figure(figsize=(15,15))
plt.plot(image4.histogram())

# Conversion
image4_gris = image4.convert('L') # 'L': Niveaux de gris, 'RGB', 'RGBA' et 'HSV' pour les couleurs (3-channel or 4-channel images)
"""
3-channel images:
    RGB images—each pixel is represented by three-tuple (r, g, b) values, representing red, green, and blue channel color values for every pixel.
    HSV images—each pixel is represented by three-tuple (h, s, v) values, representing hue (color), saturation (colorfulness—how much the color is mixed with white), and value (brightness—how much the color is mixed with black) channel color values for every pixel. The HSV model describes colors in a similar manner to how the human eye tends to perceive colors. 
4-channel images:
    Four-channel images; for example, RGBA images—each pixel is represented by three-tuple (r, g, b, α) values, the last channel representing the transparency.
"""
image4_gris.show()
image4_gris.save('test_convert_gris.bmp')
```

```python
"""===============================================
    skimage
==============================================="""
# pip install scikit-image
# pip install numpy==1.19  # pour résoudre le problème de dépendance
# ou
# apt install python3-skimage
import skimage
image5 = skimage.io.imread('test.jpeg')
print('Dimensions de l\'image: ', image5.shape)
print('Type du tableau: ', type(image5)) # le tableau image3 est du type 'numpy.ndarray'

fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(15,7), subplot_kw={'xticks': [], 'yticks': []})
axes.flat[0].set_title('Originale', size=15, color='red')
axes.flat[0].imshow(image5)
# Conversion en d'autre mode de couleurs
# https://scikit-image.org/docs/dev/api/skimage.color.html
image5_gris = skimage.color.rgb2gray(image5)
axes.flat[1].set_title('Gris', size=15, color='red')
axes.flat[1].imshow(image5_gris)
image5_hsv = skimage.color.rgb2hsv(image5)
axes.flat[2].set_title('HSV', size=15, color='red')
axes.flat[2].imshow(image5_hsv)
image5_yuv = skimage.color.rgb2yuv(image5)
axes.flat[3].set_title('YUV', size=15, color='red')
axes.flat[3].imshow(image5_yuv)
plt.tight_layout()
plt.show()
```
